```
import math

c1,c2,c3=[50],[25],[0]
t = 0.1
kf = 0.025
kb = 0.01
delta_t = 0.1
threshold = 0.01

for i in range(1,100):
  rate_c1 = kb * c3[i-1] - kf * c1[i-1] * c2[i-1]
  rate_c2 = kb * c3[i-1] - kf * c1[i-1] * c2[i-1]
  rate_c3 = 2 * kf * c1[i-1] * c2[i-1] - 2 * kb * c3[i-1]

  new_c1 = c1[i-1] + rate_c1 * delta_t
  new_c2 = c2[i-1] + rate_c2 * delta_t
  new_c3 = c3[i-1] + rate_c3 * delta_t

  c1.append(new_c1)
  c2.append(new_c2)
  c3.append(new_c3)

  t = t + delta_t
  print("at time ",i)
  print("c1 = ", new_c1)
  print("c2 = ", new_c2)
  print("c3 = ", new_c3),

  if(abs(c1[i-1] - c1[i]) <= threshold and abs(c2[i-1] - c2[i]) <= threshold and abs(c3[i-1] - c3[i]) <= threshold):
    print("\nEquilibriam Point\n")

```
