```
import math
import numpy as np
import matplotlib.pyplot as plt
```

```
z = [3]
u = []
x=['0']

for i in range(100):
  z_new = (z[i]**2 + 2*z[i] + 5) % 127
  u_new = z_new/127
  z.append(z_new)
  u.append(u_new)
  x.append(str(i+1))
print(z)
print(u)
print(x)
plt.bar(x, z)
plt.xlabel('label')
# frequency label
plt.ylabel('random number generated')
# plot title
#plt.ylim(0, 127)
plt.title('random number generate chart')
plt.show()
```

```
z1 = [7]
z2 = [3,5]
u = []
z = []
x=[]

for i in range(1,100):
  z1_new = (z1[i-1]**2 + 5) % 16
  z1.append(z1_new)
print(z1)
for i in range(2,100):
  z2_new = z2[i-1] + 2*z2[i-2] %17
  z2.append(z2_new)
print(z2)
for i in range(1,100):
  z_new = (z1[i] + 2*z2[i]) % 16
  z.append(z_new)
  u.append(z_new/16)
  x.append(str(i))
print(z)
print(u)
print(x)
plt.bar(x, z)
plt.xlabel('label')
# frequency label
plt.ylabel('random number generated')
# plot title
#plt.ylim(0, 127)
plt.title('random number generate chart')
plt.show()
```
