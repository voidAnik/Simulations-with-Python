```
import random
import matplotlib.pyplot as plt
import math
```

```
random.seed(0)
trials=[100,1000,5000,10000];
hits = 0;
hit_x=[]
hit_y=[]
miss_x=[]
miss_y=[]
pi_values=[]
for n in trials:

  for i in range(n):
    x = random.uniform(-1,1)
    y = random.uniform(-1,1)
    if((((x-0)**2)+((y-0)**2))**0.5 <= 1):
      hits=hits+1
      hit_x.append(x)
      hit_y.append(y)
    else:
      miss_x.append(x)
      miss_y.append(y)

  pi = 4*(hits/n)
  hits=0;
  pi_values.append(pi)
  print(pi)
  plt.scatter(hit_x, hit_y, color="red", label="Within circle")
  plt.scatter(miss_x, miss_y, color="green", label="Outside Cicle")
  
  #x=['100','1000','5000','10000']
  plt.ylabel('Y-label')
  plt.xlabel('X-label')
  plt.title("For trials of n = " + str(n))
  plt.legend()
  plt.show()

x=['100','1000','5000','10000']
plt.bar(x, pi_values) #pi_value
plt.xlabel('number of trials') 
  # frequency label 
plt.ylabel('value of PI') 
  # plot title 
plt.ylim(0,5)
plt.title('PI values') 
plt.show()
```
